```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que 'df_productos' es tu DataFrame que contiene la columna de productos

# Calcular el rango intercuartílico
q1 = df_productos['precio'].quantile(0.25)
q3 = df_productos['precio'].quantile(0.75)
iqr = q3 - q1

# Calcular los límites para identificar outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Filtrar los datos para encontrar outliers
outliers = df_productos[(df_productos['precio'] < lower_bound) | (df_productos['precio'] > upper_bound)]

# Visualizar los outliers mediante un diagrama de caja
plt.figure(figsize=(8, 6))
plt.boxplot(df_productos['precio'])
plt.title('Diagrama de Caja de Precios de Productos')
plt.ylabel('Precio')
plt.show()

# Mostrar los outliers encontrados
print("Outliers en la columna 'precio' de productos:")
print(outliers)
```

# Mostrar por pantalla el porcentaje de personas que
# sobrevivieron y murieron.

```python
# Contar el número de 1s y 0s en la columna Survived
conteo_sobrevivientes = df_titanic["Survived"].value_counts()

# Calcular los porcentajes
porcentajes_sobrevivientes = conteo_sobrevivientes / len(df_titanic) * 100

# Crear un DataFrame con los resultados
df_tasas_sobrevivencia = pd.DataFrame({
    "Sobrevivió": ["No", "Sí"],
    "Conteo": conteo_sobrevivientes,
    "Porcentaje": porcentajes_sobrevivientes,
})

# Generar la tabla HTML con colores diferentes para cada fila
tabla_html = df_tasas_sobrevivencia.style.set_table_styles([
    {"selector": "tr:nth-child(even)", "props": [("background-color", "#f5f5f5")]},
    {"selector": "tr:nth-child(odd)", "props": [("background-color", "#e0e0e0")]},
])

# Mostrar la tabla HTML
display(tabla_html)

```

# Mostrar por pantalla el porcentaje de personas que
# sobrevivieron en cada clase

```python

# Agregar etiquetas descriptivas a los resultados
conteo_sobrevivientes["No pudieron sobrevivir"] = conteo_sobrevivientes[0]
conteo_sobrevivientes["Sobrevivieron"] = conteo_sobrevivientes[1]

# Imprimir los resultados
print(conteo_sobrevivientes.to_string())
```

```python

# Filtrar para pasajeros que sobrevivieron
pasajeros_sobrevivientes = df_titanic[df_titanic["Survived"] == 1]

# Agrupar por Pclass y contar el número de pasajeros
conteos_sobrevivientes_por_pclass = pasajeros_sobrevivientes.groupby("Pclass").size()

# Calcular el porcentaje de sobrevivientes para cada Pclass
porcentajes_sobrevivientes_por_pclass = (conteos_sobrevivientes_por_pclass / len(pasajeros_sobrevivientes)) * 100


# Crear un DataFrame con los resultados
df_tasas_supervivencia = pd.DataFrame({
    "Pclass": conteos_sobrevivientes_por_pclass.index,
    "Recuento": conteos_sobrevivientes_por_pclass.values,
    "Porcentaje": porcentajes_sobrevivientes_por_pclass.values,
})

# ver los resultados
print(df_tasas_supervivencia.to_string())
```

# #Mostrar por pantalla la edad media de las mujeres que viajaban en cada clase.

```python
# Filtrar por pasajeras
mujeres = df_titanic[df_titanic["Sex"] == "female"]

# Agrupar por Pclass y calcular la edad media
edad_media_mujeres_pclass = mujeres.groupby("Pclass")["Age"].mean()

# Imprimir los resultados
for pclass, edad_media in edad_media_mujeres_pclass.items():
    print(f"Edad media de las mujeres en Pclass {pclass}: {edad_media:.2f}")
```

```python
#df_Titanic = df_titanic
#df_columnas_seleccionadas = df_titanic[df_titanic["Es menor de edad"] == True][["Es menor de edad", "Name","Age","Pclass"]]
#df_columnas_seleccionadas

import pandas as pd

# Cargar el conjunto de datos
df_titanic = pd.read_csv("/content/Titanic.csv")

# Añadir una nueva columna booleana para menores de edad
df_titanic["Es menor de edad"] = df_titanic["Age"] < 18

# Seleccionar las columnas relevantes para menores de edad
df_columnas_seleccionadas = df_titanic[df_titanic["Es menor de edad"] == True][["Es menor de edad", "Name","Age","Pclass"]]

# Contar los menores de edad por Pclass
conteo_menores_por_pclass = df_columnas_seleccionadas.groupby("Pclass").size()

# Calcular el porcentaje de menores de edad por Pclass
porcentaje_menores_por_pclass = (conteo_menores_por_pclass / len(df_columnas_seleccionadas)) * 100

# Crear un DataFrame con los resultados
df_menores_por_pclass = pd.DataFrame({
    "Pclass": conteo_menores_por_pclass.index,
    "Conteo": conteo_menores_por_pclass.values,
    "Porcentaje": porcentaje_menores_por_pclass.values,
})

# Imprimir el DataFrame
print(df_menores_por_pclass.to_string())
```
